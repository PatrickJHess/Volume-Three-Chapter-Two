### The `calendar` module.

The `calendar` module, frequently used for generating monthly or yearly calendars, also serves a crucial function in calculating accrued interest by identifying the last day of a month. This is relevant for bonds that mature on the last day of a month, as all payment dates will similarly fall on the last day of a month.

The `monthrange()` method:

1. accepts the **year** and **month** as integer inputs;  
2. returns a tuple containing:  
   * the day of the week for the first day of the month (with Monday represented as zero);  
   * the last calendar day of the month;

The code below shows the results for November 2025.

`````python
# The year is 2025 and the month is 11
first,last=calendar.monthrange(2025,11)
display(f'Day Of Week for First Day is {first} And Last Day Of The Month is {last}')
`````

### Determine if maturity date is last day of the month
A common Pythonic convention is to use an underscore (_) as a "throwaway" variable name for values you don't need—in this case, the index for the first day of the month. The variable `isLast` is then assigned a Boolean value by checking if `maturity_date.day` is equal to the maximum number of days in that month.

`````python
maturity_date=date(2035,4,30)
_,last=calendar.monthrange(maturity_date.year,maturity_date.month)
isLast=maturity_date.day==last
display(f'Maturity date of  {maturity_date} is last day of month? {isLast}')
`````

### The `relativedelta` method of `dateutil`

The `dateutil` library's `relativedelta` method facilitates date and time calculations. By knowing either the last or next schedled-payment date, `relativedelta` can be employed to deduce the other. The method accepts arguments specifying time intervals in years, months, days, and so forth. For instance, a semi-annual bond would require assigning a value of six to the `months` argument.

The cell below demonstrates the use of `relativedelta` by adding and subtracting six months from `maturity_date`. It's important to note that this operation disregards the month-end, effectively adding or subtracting six months minus one day.


`````python
# Assign delta_time the value returned by relativedelta: months=6
delta_time=relativedelta(months=6)
# Add six months to the maturity_ date
display(f'Six Months After Maturity Date {maturity_date+delta_time}')
# Subtract six months from the maturity_date
display(f'Six Months Before Maturity Date {maturity_date-delta_time}')
`````

## **Pandas `Timestamp`**
Python's standard `datetime` and `date` objects are suitable for handling individual or small numbers of dates. Pandas `Timestamp` objects are designed for managing large collections of dates simultaneously. A Pandas `Timestamp` can be treated like a `datetime` but offers greatly enhanced functionality. This chapter will demonstrate how to utilize Pandas `DatetimeIndex`, `DateOffset`, and `offsets.MonthEnd` to efficiently generate sequences of dates needed for determining the scheduled and actual bond payment dates.

Copy this code to create a `Timestamp`

```py
import pandas as pd
pd.Timestamp(2025,1,1)
```

### Pandas `DatetimeIndex`
Two methods are used to create indexes starting on February 29, 2024\. The examples illustrate how the choice of offset affects the index when adding months.

1. **Standard Offset:** Uses `pd.DateOffset(months=i)`, which does *not* automatically adjust for month ends (e.g., adding one month to Feb 29 yields Mar 29).  
2. **Month End Accounted:** Uses `pd.offsets.MonthEnd(i)`, which ensures the date remains the last day of the month after the offset (e.g., adding one month to Feb 29 yields Mar 31).

Copy and run the code below to see the result:

```py
import pandas as pd
from datetime import date

start = pd.Timestamp(date(2024, 2, 29))

# Standard offset - does not account for month end
index_standard = pd.DatetimeIndex(start + pd.DateOffset(months=i) for i in range(5))
# Note: The original code had a typo (display(datime_index)) which is fixed here.
# For demonstration purposes, we will display the standard index as well.

# Account for month end
index_month_end = pd.DatetimeIndex(start + pd.offsets.MonthEnd(i) for i in range(5))

print(f'Standard DateOffset: {index_standard}')
print(f'Accounted For Month End: {index_month_end}')
```
